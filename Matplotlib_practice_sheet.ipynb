# Matplotlib Subplots for Mutiple Figures

```python
#Subplot function takes 3 arguments
```

```python
#plot 1
x = np.array([0,1,2,3])
y = np.array([3,6,1,10])

plt.subplot(1,2,1) # first arg = row, 2nd arg = column, 3rd arg= diagram number
plt.plot(x,y)

#plot 2
x2 = np.array([0,1,2,3])
y2 = np.array([10,20,30,40])

plt.subplot(1,2,2)
plt.plot(x2,y2)
```

## creating scatter plots

```python
x = abs(np.random.randn(10) * 100 / 10)
y = abs(np.random.randn(10) * 100 / 10)
```

```python
plt.scatter(x, y, color='r')
plt.show
```

```python
plt.scatter(x, y, color='r', s=100)
plt.show
```

```python
#Changing sizes of scatter plot customly
```

```python
sizes = abs(np.random.randn(10) * 200 )
plt.scatter(x,y, c='r', s=sizes)
plt.show()
```
